
## Aggregrate function 

```python
from pyspark.sql.functions import sum, mean, median, mode,max,min,avg
```

```python
dff.select(mode('amt').alias("mode")).show()
```

```python
dff.select(mode('amt').alias("mode")).show()
```

```python

# state wise sales amt

dff.groupby('state').sum('amt').sort('state').show()
```

```python
# state and city wise average sales 

dff.groupby("state","city").agg(
    avg("amt").alias("average"),
    sum("amt").alias("sum of the sales"),
    min("amt").alias("min of the sales"),
    max("amt").alias("max of the sales")
).orderBy("state","city").show()
```

```python
from  pyspark.sql import Window
from pyspark.sql.functions import lag, lead, rank,year, when, lit,concat,to_date
```

```python
 win_funct = Window.partitionBy("state")\
             .orderBy("city")\
             .rowsBetween(-2,Window.currentRow)

```

```python
win_fun_2 = Window.partitionBy("state").orderBy("sum")
```

```python
txn_df_new = dff.groupby("state","city").agg(
    sum("amt").alias("sum")
     ).orderBy("state","city")

txn_df_new.withColumn("lag",lag(col("sum")).over(win_fun_2)).show()
```

### how to modify data frame column 

```python
# rename the col amt -> amount
# change the type of the cid from string to int 
# add new column as year ( derive from the txn_date)
# add new trans_cat
# if amount < 20 T1
# if  amount < 50 T2 
# otherwise T3

```

```python
dff.withColumnRenamed("amt","amount")\
   .withColumn("cid",col("cid").cast(IntegerType()))\
   .withColumn("year",year(col("txn_dt")))\
   .withColumn("trans_cat",when(col("amount") < 20, "T1")
                        .when(col("amount")< 50, "T2")
                        .otherwise("T3")).show(10)
```

```python


data_list =     [("Ravi", "28", "1", "2002"),

                 ("Abdul", "23", "5", "81"),  # 1981 

                 ("John", "12", "12", "6"),  # 2006

                 ("Rosy", "7", "8", "63"),  # 1963

                 ("Abdul", "23", "5", "81"),

                 ("Vijay", "2", "5", "21")]  # 2021
 
 
raw_df = spark.createDataFrame(data_list).toDF("name","day","month","year")

 
```

```python
raw_df.show()
```

```python
raw_df.printSchema()
```

```python
# 01. Get year in yyyy

#     if year < 27 => year+2000

#     if year < 100 => year+1900

#     else year
 
# 02. concat month,day and year with some seperator ("-") : new col name : doj

#     "28-1-2002"

#    concat(col("day"),lit("-"),col("month"),lit("-"),col("year"))
 
# 03. convert doj "28-1-2002" from string type into dateType :
 
# 04. drop three col : day, month and year
 
# 05. drop duplicate records


raw_df.withColumn("day",col("day").cast(IntegerType()))\
      .withColumn("month",col("month").cast(IntegerType()))\
      .withColumn("year",col("year").cast(IntegerType()))\
    .printSchema()
```

```python
raw_df.printSchema()
```

```python
final_df_data = raw_df.withColumn("year",col("year").cast(IntegerType()))\
     .withColumn("year",when(col("year")<27,lit(2000)+col("year"))
                        .when(col("year")<100,lit(1900)+col("year"))
                         .otherwise(col("year")))\
     .withColumn("doj",concat(col("day"),lit("-"),col("month"),lit("-"),col("year")))\
     .withColumn("doj", to_date(col("doj"), "d-M-y"))\
      .drop("day","month","year")\
      .dropDuplicates()\

```
