```python
rdd1 = sc.textFile("/home/hadoop/Downloads/custs_without_header.csv")
```

```python
rdd2 = rdd1.map(lambda rec : rec.split(','))
rdd2.take(10)
```

```python
# get the record where age is > 50

rdd3 = rdd2.filter(lambda lst : int(lst[3]) > 50)
rdd3.take(10)
```

```python
rdd4 = rdd2.filter(lambda lst : len(lst[4]) == 0)
rdd4.take(10)
```

```python
rdd5 = rdd2.filter(lambda lst : lst[4] in ['Pilot','Carpenter','Lawyer'] and int(lst[3]) > 50)
rdd5.take(10)
```

```python
rdd6 = rdd2.filter(lambda lst : lst[4] == 'Pilot' and int(lst[3]) > 50)
rdd6.take(10)
```
